CODE FOR SPLICING OUR DATA

```python
from pydub import AudioSegment
import pandas as pd
import os


# Example usage
audio_dir = '/content/drive/My Drive/DT2_2023/Spring 2024/Audio_models/splicing_code/full_audios/'
labels_file = '/content/drive/My Drive/DT2_2023/Spring 2024/Audio_models/splicing_code/labels/intersections/'
output_dir = '/content/drive/My Drive/DT2_2023/Spring 2024/Audio_models/audio_input/mass_intersections_3'
swallow = 0; # 1 is for swallow audio, 0 is for nonswallow audio
clip_duration=2000

num_swallow_audios = 0;

for filename in os.listdir(labels_file):
  if "noswall" in filename:
    swallow = 0;
    start = 8;
    print(filename)
  else:
    swallow = 1;
    start = 0;

  audio_name = filename[start];
  dashes = 0;

  for i in range(start+1, len(filename) - 5):
    if filename[i] == '-':
      dashes = dashes + 1;
    if dashes == 4:
      break;
    audio_name = audio_name + filename[i];


  print(audio_name)
  # Load audio file
  audio = AudioSegment.from_wav(audio_dir + audio_name + '.wav')
  audio = audio.set_channels(1)

  # Read labels file into pandas DataFrame
  labels_df = pd.read_csv(labels_file + filename, delimiter='\t', header=None, names=['start', 'end', 'label'])

  # Convert time to milliseconds
  labels_df['start_ms'] = (labels_df['start'] * 1000).astype(int)
  labels_df['end_ms'] = (labels_df['end'] * 1000).astype(int)

  # Splice audio clips
  for idx, row in labels_df.iterrows():
      for i in range(0, clip_duration+1, 100):
          if swallow:
            start_ms = row['start_ms'] - i;  # Adjust start time
          else:
            start_ms = row['start_ms'];
          end_ms = start_ms + clip_duration      # Adjust end time

          if start_ms < 0:
            start_ms = 0
          if end_ms > len(audio):
            end_ms = len(audio)
          if (swallow == 0) and (end_ms > row['end_ms']):
            break;
          if (end_ms - start_ms) < clip_duration:
            break;

          clip = audio[start_ms:end_ms]
          if swallow:
            output_file = f"{output_dir}/class_1/{audio_name}_splice_{idx}_{int(i/100)}.wav"
          else:
            output_file = f"{output_dir}/class_0/{audio_name}_splice_{idx}_{int(i/100)}.wav"
          clip.export(output_file, format="wav")
          num_swallow_audios = num_swallow_audios + 1;

print(num_swallow_audios)
```
