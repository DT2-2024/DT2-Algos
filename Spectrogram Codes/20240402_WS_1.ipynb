```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import wave
import pylab
from pathlib import Path
from scipy import signal
from scipy.io import wavfile
from sklearn.metrics import confusion_matrix
import itertools

# Set paths to input and output data
INPUT_DIR = '/content/drive/My Drive/DT2_2023/Spring 2024/Audio_models/audio_input/mass_intersections_2'
OUTPUT_DIR = '/content/drive/My Drive/DT2_2023/Spring 2024/Audio_models/spectro_output/WS_1_mass_intersectections_2_20240415'
```

```python
# Utility function to get sound and frame rate info
def get_wav_info(wav_file):
    wav = wave.open(wav_file, 'r')
    frames = wav.readframes(-1)
    sound_info = pylab.frombuffer(frames, 'int16')
    frame_rate = wav.getframerate()
    wav.close()
    return sound_info, frame_rate

for i in range(0, 2):
  if not os.path.exists(os.path.join(OUTPUT_DIR, 'class_' + str(i))):
      os.mkdir(os.path.join(OUTPUT_DIR, 'class_' + str(i)))

  temp_INPUT_DIR = INPUT_DIR + '/class_' + str(i)

  for filename in os.listdir(temp_INPUT_DIR):
      if "wav" in filename:
          file_path = os.path.join(temp_INPUT_DIR, filename)
          file_stem = Path(file_path).stem
          target_dir = f'class_{str(i)}'
          dist_dir = os.path.join(OUTPUT_DIR, target_dir)
          file_dist_path = os.path.join(dist_dir, file_stem)
          if not os.path.exists(file_dist_path + '.png'):
              if not os.path.exists(dist_dir):
                  os.mkdir(dist_dir)
              file_stem = Path(file_path).stem
              sound_info, frame_rate = get_wav_info(file_path)
              pylab.specgram(sound_info, Fs=frame_rate, NFFT=2048)
              pylab.savefig(f'{file_dist_path}.png')
              pylab.close()
```
